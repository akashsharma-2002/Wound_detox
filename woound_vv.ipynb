```python

import cv2
from ultralytics import YOLO
import time
import pytesseract
import easyocr
import matplotlib.pyplot as pl
import ssl
import glob
import random
from gemi import generate_and_format_markdown,analyze_wound
import os
# Specify the path to the tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/bin/tesseract'

# Initialize the model
model = YOLO('best.pt')

# Open the camera
cap = cv2.VideoCapture(0)

# Get the start time
start_time = time.time()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Run the model on the frame
    results = model(frame, show=True, conf=0.4, save=True)

    # If 10 seconds have passed, break the loop
    if time.time() - start_time > 10:
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()

# converting image to text
ssl._create_default_https_context = ssl._create_unverified_context

# Get the latest file in the directory
list_of_files = glob.glob('runs/detect/predict*/image0.jpg')  # * means all if need specific format then *.csv
latest_file = max(list_of_files, key=os.path.getctime)

img=cv2.imread(latest_file)
reader=easyocr.Reader(['en'],gpu=True)
results=reader.readtext(img)

letter_list = []  # Initialize an empty list to store the letters

for result in results:
    text = result[1]  # Get the recognized text from the result tuple
    first_letter = text[0].lower()  # Get the first letter of the text and convert it to lowercase
    letter_list.append(first_letter)  # Add the first letter to the list


for i in letter_list:
    print(i)
    if i=="b":
        # latest_image_path = max(glob.glob('runs/detect/predict*/image0.jpg'), key=os.path.getmtime, default=None)
        # if latest_image_path:
            
        #     markdown_content = analyze_wound(latest_image_path)
        #     print(display(markdown_content))
        # else:
        #     print("No image found for analysis.")
            
        c=generate_and_format_markdown("Bruise")
        print(display(c))
    elif(i=="l"):
        # latest_image_path = max(glob.glob('runs/detect/predict*/image0.jpg'), key=os.path.getmtime, default=None)
        # if latest_image_path:
        #     markdown_content = analyze_wound(latest_image_path)
        #     print(display(markdown_content))
        # else:
        #     print("No image found for analysis.")
        
        c=generate_and_format_markdown("Laceration")
        print(display(c))
    elif(i=="c"):
        # latest_image_path = max(glob.glob('runs/detect/predict*/image0.jpg'), key=os.path.getmtime, default=None)
        # if latest_image_path:
            
        #     markdown_content = analyze_wound(latest_image_path)
        #     print(display(markdown_content))
        # else:
        #     print("No image found for analysis.")
        
        c=generate_and_format_markdown("cut wound")
        print(display(c))
    elif(i=="s"):
        # latest_image_path = max(glob.glob('runs/detect/predict*/image0.jpg'), key=os.path.getmtime, default=None)
        # if latest_image_path:
            
        #     markdown_content = analyze_wound(latest_image_path)
        #     print(display(markdown_content))
        # else:
        #     print("No image found for analysis.")
        
        c=generate_and_format_markdown("stab wound")
        print(display(c))
    elif(i=="a"):
        # latest_image_path = max(glob.glob('runs/detect/predict*/image0.jpg'), key=os.path.getmtime, default=None)
        # if latest_image_path:
            
        #     markdown_content = analyze_wound(latest_image_path)
        #     print(display(markdown_content))
        # else:
        #     print("No image found for analysis.")

        c=generate_and_format_markdown("Abrasion")
        print(display(c))
    
    

```
